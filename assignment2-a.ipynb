### Assignment 2: Classification using Deep neural network
1. Multiclass classification using Deep Neural Networks: Example: Use the OCR letter 
recognition dataset https://archive.ics.uci.edu/ml/datasets/letter+recognition 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix as cm1
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("letter-data/letter-recognition.data", sep=",")

df
```

```python
X = df.iloc[:, 1:17]
Y = df.iloc[:, 0]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
Y_encoded = le.fit_transform(Y)
```

```python
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(
    X, Y_encoded, test_size=0.2, random_state=42
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_validation = scaler.transform(X_validation)
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    max_iter=300,
    activation='relu',
    solver='adam',
    early_stopping=True,
    verbose=1,
    random_state=42
)
```

```python
mlp.fit(X_train, Y_train)
```

```python
Y_pred = mlp.predict(X_validation)
Y_pred_labels = le.inverse_transform(Y_pred)

print("Accuracy:", accuracy_score(le.inverse_transform(Y_validation), Y_pred_labels))

print(classification_report(le.inverse_transform(Y_validation), Y_pred_labels))
```
